```python
#condensons les données
# On définit comment regrouper chaque colonne
logiciel_agg = {
    'Open': 'first',   # Premier prix de la journée
    'High': 'max',     # Plus haut prix de la journée
    'Low': 'min',      # Plus bas prix de la journée
    'Close': 'last',   # Dernier prix de la journée
    'Volume': 'sum'    # Somme de tous les échanges du jour
}

# 'D' pour Daily (Journalier), tu peux mettre 'M' pour Monthly (Mensuel)
df_day = df.resample('D').agg(logiciel_agg)

print(df_day.head())
print(f"Nouvelle taille : {df_day.shape}")
```

## Modélisation Statistique (ARIMA / SARIMA)

```python
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
# On peut prendre 80% pour training, 20% pour test
train_size = int(len(df_day)*0.8)
train, test = df_day['Close'][:train_size], df_day['Close'][train_size:]

```

```python
# ARIMA simple
model = ARIMA(train, order=(5,1,0))  # p,d,q à ajuster via ACF/PACF
model_fit = model.fit()
preds = model_fit.forecast(steps=len(test))

```

```python
# RMSE
rmse = np.sqrt(mean_squared_error(test, preds))
print(f"RMSE ARIMA: {rmse}")

plt.figure(figsize=(12,5))
plt.plot(train.index, train, label='Train')
plt.plot(test.index, test, label='Test')
plt.plot(test.index, preds, label='ARIMA Prediction', color='red')
plt.legend()
plt.show()
```

```python
import numpy as np

# 1. Stabiliser la tendance (Logarithme du prix)
df_day['Log_Price'] = np.log(df_day['Close'])

# 2. Rendre la série stationnaire (Rendements logarithmiques)
# Cela correspond à la variation en % d'un jour à l'autre
df_day['Log_Return'] = df_day['Log_Price'].diff().dropna()

# Visualisation
plt.figure(figsize=(12,5))
plt.plot(df_day.index, df_day['Log_Price'], color='green')
plt.title('Prix du Bitcoin en Échelle Logarithmique')
plt.ylabel('Log(Prix)')
plt.show()
```

```python
#Entraînement sur le Log-Price

# 1. Transformation Logarithmique (on utilise les données journalières df_day)
df_day['Log_Price'] = np.log(df_day['Close'])

# 2. Split Train/Test sur le Log
train_size = int(len(df_day) * 0.8)
train_log, test_log = df_day['Log_Price'][:train_size], df_day['Log_Price'][train_size:]
test_actual = df_day['Close'][train_size:] # Pour la comparaison finale en $

# 3. Modèle ARIMA sur le Log
# On garde (5,1,0) : le '1' fera la différence sur le Log, ce qui revient à étudier les rendements.
model_log = ARIMA(train_log, order=(5,1,0))
model_fit_log = model_log.fit()

# 4. Prédiction (toujours en log)
preds_log = model_fit_log.forecast(steps=len(test_log))

# 5. Conversion inverse : Repasser du Log au Prix Réel ($)
preds_final = np.exp(preds_log)

# 6. Calcul du nouveau RMSE
rmse_log_model = np.sqrt(mean_squared_error(test_actual, preds_final))
print(f"Ancien RMSE (brut): 51531")
print(f"Nouveau RMSE (via Log): {rmse_log_model}")
```

```python
plt.figure(figsize=(12,6))
plt.plot(test_actual.index, test_actual, label='Prix Réel (Test)', color='orange')
plt.plot(test_actual.index, preds_final, label='Prédiction ARIMA (via Log)', color='red')
plt.title('Comparaison : Réel vs Prédiction (Espace Dollars)')
plt.legend()
plt.show()
```

```python
## Le Rolling Forecast 
history = [x for x in train_log]
predictions = []

# On boucle sur chaque point du dataset de test (limite à 30 points pour tester)
for t in range(len(test_log[:30])):
    model = ARIMA(history, order=(5,1,0))
    model_fit = model.fit()
    
    # Prédit seulement le point suivant
    output = model_fit.forecast()
    yhat = output[0]
    predictions.append(np.exp(yhat))
    
    # On ajoute la VRAIE valeur à l'historique pour la prochaine itération
    obs = test_log.iloc[t]
    history.append(obs)

# Visualisation des 30 premiers jours de test
plt.plot(test_actual.values[:30], label='Réel')
plt.plot(predictions, color='red', label='Rolling ARIMA')
plt.legend()
plt.show()
```

1. Analyse du succès : Le "Rolling Forecast" 
C'est le jour et la nuit par rapport aux graphiques précédents (Images 3 et 4).

Réactivité : En donnant au modèle la vraie valeur de la veille pour prédire demain, on as permis à l'ARIMA de se recalibrer.
La ligne rouge "colle" au prix réel car elle utilise la mémoire immédiate.

Le décalage (Lag) : Si on regardes de très près, on verras que la ligne rouge semble souvent "suivre" la bleue avec un petit temps de retard.
C'est typique d'ARIMA sur des marchés volatils : il prédit souvent que "demain sera un peu comme aujourd'hui".

2. Comparaison des approches
ARIMA Brut Ligne plate  Nulle pour le Bitcoin (trop de croissance)
ARIMA via Log  Ligne plate (plus stable) Faible sur le long terme sans mise à jour.
Rolling ARIMA Courbe dynamique Réaliste pour de la prédiction à court terme ($J+1$).

3. Synthèse de la décomposition 
les graphiques de décomposition ont servi de fondation :

Ils ont prouvé que le Bitcoin est dominé par sa Tendance (Trend) et non par des cycles répétitifs (Saisonnalité).

C'est pour cela que le modèle Rolling fonctionne : il "surf" sur la tendance actuelle plutôt que de chercher un cycle fantôme.

## ML non séquentiel : XGBoost

Contrairement à ARIMA qui tente de modéliser la structure mathématique de la série, XGBoost traite cela comme un problème de régression classique : "Connaissant les 3 derniers jours, quel est le prix aujourd'hui ?".

```python
import xgboost as xgb
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

# 1. Préparation des Features (Ingénierie des données)
# On s'assure de travailler sur une copie pour ne pas modifier l'original
df_ml = df_day.copy()

# Lags (Valeurs passées)
df_ml['t-1'] = df_ml['Close'].shift(1)
df_ml['t-2'] = df_ml['Close'].shift(2)
df_ml['t-3'] = df_ml['Close'].shift(3)

# Indicateurs Techniques
df_ml['SMA_7'] = df_ml['Close'].rolling(window=7).mean()   # Tendance court terme
df_ml['SMA_30'] = df_ml['Close'].rolling(window=30).mean() # Tendance long terme
df_ml['Std_7'] = df_ml['Close'].rolling(window=7).std()    # Volatilité

# Features de Calendrier
df_ml['DayOfWeek'] = df_ml.index.dayofweek
df_ml['Month'] = df_ml.index.month

# Nettoyage des lignes avec des NaN (dues au shifting et rolling)
df_ml.dropna(inplace=True)

# 2. Définition des variables X (Entrées) et y (Cible)
features = ['t-1', 't-2', 't-3', 'SMA_7', 'SMA_30', 'Std_7', 'DayOfWeek', 'Month']
X = df_ml[features]
y = df_ml['Close']

# 3. Split Train/Test (80% / 20%) - SANS SHUFFLE pour respecter la chronologie
split_idx = int(len(df_ml) * 0.8)
X_train, X_test = X.iloc[:split_idx], X.iloc[split_idx:]
y_train, y_test = y.iloc[:split_idx], y.iloc[split_idx:]

# 4. Création et Entraînement du modèle XGBoost
# On utilise des hyperparamètres prudents pour éviter le sur-apprentissage (overfitting)
xgb_model = xgb.XGBRegressor(
    n_estimators=500, 
    learning_rate=0.01, 
    max_depth=5, 
    subsample=0.8, 
    colsample_bytree=0.8,
    objective='reg:squarederror'
)

xgb_model.fit(X_train, y_train)

# 5. Prédictions
preds_xgb = xgb_model.predict(X_test)
preds_series = pd.Series(preds_xgb, index=y_test.index)

# 6. Évaluation
rmse_xgb = np.sqrt(mean_squared_error(y_test, preds_xgb))
print(f"--- RÉSULTATS ---")
print(f"RMSE XGBoost: {rmse_xgb:.2f} USD")

# 7. Visualisation
plt.figure(figsize=(14, 7))
plt.plot(y_test.index, y_test, label='Prix Réel (Test)', color='blue', alpha=0.6)
plt.plot(preds_series.index, preds_series, label='Prédiction XGBoost', color='red', linestyle='--')
plt.title('Bitcoin Price Prediction: XGBoost avec Features Techniques')
plt.xlabel('Date')
plt.ylabel('Prix USD')
plt.legend()
plt.grid(True)
plt.show()

# Bonus : Importance des variables
xgb.plot_importance(xgb_model)
plt.title('Importance des Features dans la décision du modèle')
plt.show()
```

1. Analyse du graphique de prédiction
La Phase de Mimétisme (Mai 2023 - Janvier 2024) : Au début de la période de test, la ligne rouge (XGBoost) colle presque parfaitement à la ligne bleue. C'est normal : le prix se situe dans des zones de valeurs déjà explorées durant l'entraînement.

Le "Plafond de Verre" (Après Janvier 2024) : Dès que le prix du Bitcoin explose au-dessus de 60 000 USD, le modèle XGBoost s'arrête net. Il prédit une ligne de bruit horizontale autour de 62k.

Pourquoi ? Un modèle basé sur des arbres (XGBoost) ne peut pas prédire une valeur supérieure au maximum de ses données d'entraînement. C'est le problème de l'incapacité d'extrapolation. Pour lui, un prix de 100k est physiquement "impossible" car il ne l'a jamais vu.

2. Analyse de l'Importance des Features (Bas)
C'est ici que l'on vois comment le modèle "réfléchit" :

t-1 (Score : 4583.0) : C'est de loin la variable la plus puissante. Le modèle a appris que le meilleur indicateur du prix de demain est tout simplement le prix d'aujourd'hui.

Std_7 (Score : 1841.0) : Très intéressant ! la feature de volatilité arrive en deuxième position. Cela prouve que le modèle ajuste ses prédictions en fonction de la nervosité du marché.

SMA_7 & SMA_30 : Les moyennes mobiles sont également très utiles pour donner le contexte de la tendance au modèle.

Le Temps (Month, DayOfWeek) : Ces variables ont les scores les plus bas. Pour le Bitcoin, le jour de la semaine semble beaucoup moins important que la dynamique du prix lui-même.

### prédire le rendement (la variation) plutôt que le prix brut

```python
import pandas as pd
import numpy as np
import xgboost as xgb
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

# ==========================================
# 1. PRÉPARATION DES DONNÉES (FEATURES)
# ==========================================
# On suppose que df_day est ton dataframe journalier avec Date en index
df_ml = df_day.copy()

# Cible : On prédit la VARIATION (le Delta) pour éviter le plafonnement
df_ml['Delta'] = df_ml['Close'].diff()

# Features de retard (Lags) sur le prix
df_ml['t-1'] = df_ml['Close'].shift(1)
df_ml['t-2'] = df_ml['Close'].shift(2)

# Indicateurs Techniques
df_ml['SMA_7'] = df_ml['Close'].rolling(window=7).mean()   # Moyenne mobile 7j
df_ml['Std_7'] = df_ml['Close'].rolling(window=7).std()    # Volatilité 7j
df_ml['Rolling_Delta'] = df_ml['Delta'].rolling(window=3).mean() # Tendance du delta

# Calendrier
df_ml['DayOfWeek'] = df_ml.index.dayofweek

# Nettoyage des NaN
df_ml.dropna(inplace=True)

# ==========================================
# 2. SPLIT ET DÉFINITION X, y
# ==========================================
features = ['t-1', 't-2', 'SMA_7', 'Std_7', 'Rolling_Delta', 'DayOfWeek']
X = df_ml[features]
y = df_ml['Delta'] # On apprend à prédire le changement

# Split chronologique (80% Train, 20% Test)
split = int(len(df_ml) * 0.8)
X_train, X_test = X.iloc[:split], X.iloc[split:]
y_train, y_test = y.iloc[:split], y.iloc[split:]
actual_prices = df_ml['Close'].iloc[split:] # Prix réels pour comparer

# ==========================================
# 3. MODÈLE XGBOOST
# ==========================================
model = xgb.XGBRegressor(
    n_estimators=200,
    learning_rate=0.05,
    max_depth=5,
    objective='reg:squarederror'
)

model.fit(X_train, y_train)

# ==========================================
# 4. PRÉDICTIONS ET RECONSTRUCTION
# ==========================================
# Le modèle prédit un Delta (ex: +500$)
preds_delta = model.predict(X_test)

# Reconstruction du prix : Prix_Hier + Delta_Prédit
# On utilise .values pour éviter les problèmes d'alignement d'index
preds_price = X_test['t-1'].values + preds_delta

# ==========================================
# 5. MÉTRIQUES ET VISUALISATION
# ==========================================
# RMSE en Dollars
rmse = np.sqrt(mean_squared_error(actual_prices, preds_price))

# Directional Accuracy (Sens de la montée/descente)
# On compare le signe du vrai delta avec le signe du delta prédit
correct_direction = np.sign(y_test.values) == np.sign(preds_delta)
accuracy = np.mean(correct_direction) * 100

print(f"--- PERFORMANCE DU MODÈLE ---")
print(f"RMSE : {rmse:.2f} USD")
print(f"Précision Directionnelle : {accuracy:.2f} %")

# Graphique Final
plt.figure(figsize=(15, 7))
plt.plot(actual_prices.index, actual_prices, label='Prix Réel (Test)', color='blue', alpha=0.6)
plt.plot(actual_prices.index, preds_price, label='Prédiction XGBoost (Delta)', color='red', linestyle='--')
plt.title('Bitcoin : Prédiction via XGBoost (Méthode Delta)')
plt.legend()
plt.grid(True)
plt.show()
```

1. Analyse des Métriques (Image 8)RMSE : 1 771.71 USD 
C'est une amélioration spectaculaire ! le premier ARIMA affichait 51 531 et le premier XGBoost 26 424.
Une erreur moyenne de ~1700$ sur un actif qui en vaut 100 000 représente une erreur relative de seulement 1.7%.
C'est très robuste pour du journalier.Précision Directionnelle : 65.67%C'est le chiffre le plus important.
En finance, si tu devines la direction (hausse/baisse) plus de 55% du temps, tu as un avantage statistique.
65% est un score exceptionnel. Cela signifie que ton modèle a capturé une vraie logique dans les mouvements du Bitcoin.

2. Pourquoi ça a fonctionné 
Le graphique montre que la ligne rouge (pointillée) suit fidèlement les pics et les creux de la ligne bleue.
Adaptabilité : En prédisant le "Delta" (la variation), le modèle n'a plus peur des nouveaux records de prix.
Il se contente d'ajouter ou soustraire une valeur au prix d'hier.
Récupération de la Tendance : Contrairement à l'ARIMA qui devenait plat, XGBoost utilise ses features (SMA_7, Std_7) 
pour comprendre si le marché est en phase d'accélération ou de correction.

3. Les limites à garder en tête
Malgré ce beau graphique, attention à "l'effet miroir" :Le modèle utilise le prix de la veille ($t-1$) pour construire sa prédiction.
Si le Bitcoin chute de 10% soudainement, le modèle ne le prédira probablement pas avant que ça n'arrive, mais il s'ajustera immédiatement le lendemain.
C'est un modèle de très court terme (24h).

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from sklearn.preprocessing import MinMaxScaler


# On suppose que df_day est ton dataframe journalier avec Date en index
df_ml = df_day.copy()
# Normalisation
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df_ml[['Close']])

# Création dataset séquentiel
def create_sequences(data, seq_length=30):
    X, y = [], []
    for i in range(seq_length, len(data)):
        X.append(data[i-seq_length:i, 0])
        y.append(data[i, 0])
    return np.array(X), np.array(y)

SEQ_LENGTH = 30
X, y = create_sequences(scaled_data, SEQ_LENGTH)
X = X.reshape((X.shape[0], X.shape[1], 1))

train_size = int(0.8*len(X))
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

# LSTM
model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(SEQ_LENGTH,1)))
model.add(Dropout(0.2))
model.add(LSTM(50))
model.add(Dropout(0.2))
model.add(Dense(1))

model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(X_train, y_train, epochs=20, batch_size=32, validation_split=0.1)

# Prédiction
preds_lstm = model.predict(X_test)
preds_lstm = scaler.inverse_transform(preds_lstm)
y_test_rescaled = scaler.inverse_transform(y_test.reshape(-1,1))

rmse_lstm = np.sqrt(mean_squared_error(y_test_rescaled, preds_lstm))
print(f"RMSE LSTM: {rmse_lstm}")
```

### Prédiction "Live" (J+1)

```python
# 1. Récupérer les 30 derniers jours de ton dataset
last_30_days = df_day['Close'].values[-SEQ_LENGTH:]
last_30_days_scaled = scaler.transform(last_30_days.reshape(-1, 1))

# 2. Préparer la structure pour le modèle (Batch, Steps, Features)
X_input = last_30_days_scaled.reshape((1, SEQ_LENGTH, 1))

# 3. Prédire
prediction_scaled = model.predict(X_input)

# 4. Inverser la normalisation pour avoir le prix en $
prediction_price = scaler.inverse_transform(prediction_scaled)

last_price = df_day['Close'].iloc[-1]
diff = prediction_price[0][0] - last_price
direction = "HAUSSE" if diff > 0 else "BAISSE"

print(f"--- PRÉDICTION POUR DEMAIN ---")
print(f"Dernier prix connu : {last_price:.2f} $")
print(f"Prix prédit par LSTM : {prediction_price[0][0]:.2f} $")
print(f"Tendance prévue : {direction} ({diff:+.2f} $)")
```

```python
import requests
import pandas as pd
import plotly.graph_objects as go

# 1. Requête à ton API locale
url = "http://127.0.0.1:8000/predict?days=7"
response = requests.get(url)

if response.status_code == 200:
    data = response.json()
    df_forecast = pd.DataFrame(data['predictions'])
    
    print(f"✅ Prédictions récupérées pour la période débutant le : {data['derniere_maj']}")
    
    # 2. Affichage du tableau propre
    display(df_forecast.style.bar(subset=['variation_prevue'], color=['#ff9999', '#99ff99']))
    
    # 3. Graphique interactif avec Plotly
    fig = go.Figure()
    fig.add_trace(go.Scatter(
        x=df_forecast['jour'], 
        y=df_forecast['prix_estime'],
        mode='lines+markers',
        name='Prix Estimé BTC',
        line=dict(color='firebrick', width=4)
    ))
    
    fig.update_layout(
        title='Prévisions du Consensus (Prochains 7 Jours)',
        xaxis_title='Jours à venir (J+x)',
        yaxis_title='Prix estimé (USD)',
        template='plotly_dark'
    )
    fig.show()
else:
    print("❌ Erreur : Vérifie que la cellule de l'API est bien en train de tourner.")
```
